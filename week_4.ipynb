```python
import pandas as pd
import numpy as np
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.feature_selection import SelectFromModel
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('orange_small_churn_train_data.csv')
test_df = pd.read_csv('orange_small_churn_test_data.csv')
```

```python
# Отделим лейблы, выкинем id
labels = df['labels']
id_df = test_df['ID']
df = df.drop(['labels', 'ID'], axis=1)
labels = pd.Series(np.where(labels == -1, 0, 1))
test_df.drop(['ID'], axis=1, inplace=True)
```

```python
labels.value_counts()
```

```python
# Функция. Если количество уникальных значений больше unique, то им присваивается числа, если этих значений больше cnt. 
# Если значение меньше cnt, все такие значения объединяются.
# Если значений меньше unique, то каждому присваивается цифра. Самые многочисленные оставляем, остальные объединяем
def change_labels(column, cnt, unique):
    val = column.value_counts()
    new_val = (val >= cnt) * range(1, len(val) + 1) if len(val) > unique else (val >= 0) * range(1, len(val) + 1)
    new_my_col = column.map(new_val)
    return new_my_col


def plot_cat_unique(cat):
    plt.figure(figsize=(20, 5))
    sns.barplot(x='index', y=0, data=cat.apply(lambda x: np.unique(x).shape[0], axis=0).reset_index())
    plt.title('Количество уникальных категорий у категориальных признаков')
    plt.ylabel('Количество')
    plt.xlabel('Признак')
    plt.xticks(rotation=90);
    
def data_transform(df, cat_cnt, cat_unique):
    # Функция для dummy-преобразования. 
    # Cat_cnt - При условии количества категорий, принимающих это значение, эта категория кодируется собственным числом. 
    # cat_unique - Если количество уникальных категорий больше этого значения, то те, которые в меньшинстве, объединяются.
    df.dropna(axis=1, how='all', inplace=True)
    real = df.loc[:, :'Var190']
    real.fillna(real.mean(), inplace=True)
    real = pd.DataFrame(StandardScaler().fit_transform(real), columns=real.columns)
    # Добавим квадраты признаков
    real_sqr = real ** 2
    real_sqr.columns = real.columns+'_sqr'
    # Теперь займемся категорийными.
    cat = df.loc[:, 'Var191':]
    cat.fillna('no_value', axis=1, inplace=True)
    cat = cat.apply(lambda x: change_labels(x, cat_cnt, cat_unique), axis=0)
    dummy_cat = pd.get_dummies(data=cat, columns=cat.columns, drop_first=True)
    clear_data = pd.concat((real, real_sqr, dummy_cat), axis=1)
    
    return clear_data
```

```python
# Соединим датафреймы, чтобы произвести dummy-преобразование
concated_df = pd.concat((df, test_df), ignore_index=True)
```

```python
plot_cat_unique(df.loc[:, 'Var191':].fillna('no_value'))
```

```python
# Произведем dummy-преобразование, объединив категории, что в меньшинстве, иначе датафрейм будет слишком громоздкий и не влезет в память
dummy_data = data_transform(concated_df, 500, 30)
```

```python
# Разделим данные обратно на train & test
dummy_train = dummy_data.iloc[:39999, :]
dummy_test = dummy_data.iloc[40000:, :]
```

```python
# Теперь сделаем отбор признаков. Так как выборка несбалансирована, добавим вектор весов.
weights_vector = list(map(lambda x: (x / (labels.sum() / labels.shape[0])) if x == 1 else 1, labels))
est = XGBClassifier(n_estimators=50, max_depth=5)
est.fit(dummy_train, labels, sample_weight=weights_vector)
plt.figure(figsize=(20, 5))
plt.plot(est.feature_importances_, 'r+')
```

```python
cols_selected = dummy_train.columns[est.feature_importances_ > 0.005]
selected_train = dummy_train[cols_selected]
print('Отобрали {} признаков из {} с помощью xgb'.format(cols_selected.shape[0], dummy_train.shape[1]))
selected_test = dummy_test[cols_selected]
```

```python
# Для следующего задания: сохраним трэйн_дату в датафрейм
dummy_train.to_csv('selected_data.csv', index=False)
labels.to_csv('selected_labels.csv', index=False)
```

```python
# Обучим на части очищенных данных XGB и проверим roc_auc
X_train, X_test, y_train, y_test = train_test_split(selected_train, labels, test_size=0.2, stratify=labels)
weights_vector = list(map(lambda x: (x / (y_train.sum() / y_train.shape[0])) if x == 1 else 1, np.array(y_train)))
est_weighted = XGBClassifier(n_estimators=50, max_depth=5)
est_weighted.fit(X_train, y_train, sample_weight=weights_vector)
```

```python
predicted_proba = est_weighted.predict_proba(X_test)[:, 1]
print(metrics.roc_auc_score(y_test, predicted_proba))
print(metrics.classification_report(y_test, est_weighted.predict(X_test)))
```

Результаты не великолепные, но, в целом, по ROC-AUC они нас устраивают. Из-за больших весов на искомый класс получился низкий precision за счет высокого recall. Можно попробовать немного оптимизировать, и выкладывать на kaggle.

```python
weights_vector_full = list(map(lambda x: (x / (labels.sum() / labels.shape[0])) if x == 1 else 1, labels))
est = XGBClassifier()

params = {'max_depth': [3, 4, 5],
         'n_estimators': [40, 50, 60],
         }

xgrid = GridSearchCV(est, param_grid=params, scoring='roc_auc', iid=False, cv=3, return_train_score=False)
xgrid.fit(selected_train, labels, sample_weight=weights_vector_full)
```

```python
xgrid.best_params_
```

```python
est = XGBClassifier(max_depth=3, n_estimators=60, silent=False).fit(selected_train, labels, sample_weight=weights_vector_full)
predicted_proba = est.predict_proba(selected_test)[:, 1]
result_df = pd.DataFrame(data={'Id': id_df.values, 'result': predicted_proba})
result_df.to_csv('churn_kaggle.csv', index=False)
```
